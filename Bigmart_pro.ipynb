# Attribute Information

Item_Identifier: Unique product ID

Item_Weight: Weight of product

Item_Fat_Content: Whether the product contain low fat or not

Item_Visibility: The Percentage of total display area of all products in a store allocated to the particular product

Item_Type: The category to which the product belongs

Item_MRP: Maximum Retail Price (list price) of the product

Outlet_Identifier: Unique store ID

Outlet_Establishment_Year: The year in which store was established

Outlet_Size: The size of the store in terms of ground area covered

Outlet_Location_Type: The type of city in which the store is located

Outlet_Type: Whether the outlet is just a grocery store or some sort of supermarket

Item_Outlet_Sales: Sales of the product in the particulat store. This is the outcome variable to be predicted.

# Removing outliers and skewness from the dataset

```python
from scipy.stats import zscore
```

```python
z=np.abs(zscore(bigm))
print(z)
```

```python
threhold=3
```

```python
print(np.where(z>3))#printing all the value position which are greater than the threshold value
```

```python
z[49][2]#checking the above output result 
```

```python
bigm_s=bigm[(z<3).all(axis=1)] #removing the outliers
```

```python
print(bigm_s.shape, bigm.shape)
```

```python
#we are left with 8222 rows after removing the outliers
```

```python
#checkig the skewness after removing the outliers
bigm_s.skew()
```

```python
#removing the skewness
colllist=bigm_s.columns.values
for i in range(0,len(colllist)):
    if bigm_s.skew().loc[colllist[i]]>0.55:
        bigm_s[colllist[i]]=np.sqrt(bigm_s[colllist[i]])
```

```python
#rechecking the skewness after removing it above
bigm_s.skew()
```

```python
#almost skewness is removed except outlet_type column
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
#Finding the best Random state

max_score=0
final_rstate=0

for i in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=i)
    lm=LinearRegression() 
    lm.fit(x_train,y_train)
    lm.score(x_train,y_train)
    pred=lm.predict(x_test)
    r_score=r2_score(y_test,pred)
    print("R2 SCORE is equal to :",r_score,"at random state",i)
    
    if r_score > max_score:
        max_score=r_score
        final_rstate=i
        
        
    
    
print("The maximum r2_score is ",max_score, "at random state",final_rstate) 
```
