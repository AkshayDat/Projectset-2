# Removing outliers and skewness from the dataset

```python
from scipy.stats import zscore
z=np.abs(zscore(cov_df))
z
```

```python
threshold=3
```

```python
print(np.where(z>3))#printing all the value position which are greater than the threshold value
```

```python
z[7932][2]#checking the above output result 
```

```python
covid=cov_df[(z<3).all(axis=1)] #removing the outliers
```

```python
#checking the skewness afterremoving outliers
covid.skew()
```

```python
#removing the skewness
colllist=covid.columns.values
for i in range(0,len(colllist)):
    if covid.skew().loc[colllist[i]]>0.55:
        covid[colllist[i]]=np.sqrt(covid[colllist[i]])
```

```python
#recheking the skewness
covid.skew()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
#Finding the best Random state

max_score=0
final_rstate=0

for i in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=i)
    lm=LinearRegression() 
    lm.fit(x_train,y_train)
    lm.score(x_train,y_train)
    pred=lm.predict(x_test)
    r_score=r2_score(y_test,pred)
    print("R2 SCORE is equal to :",r_score,"at random state",i)
    
    if r_score > max_score:
        max_score=r_score
        final_rstate=i
        
        
    
    
print("The maximum r2_score is ",max_score, "at random state",final_rstate) 
```
